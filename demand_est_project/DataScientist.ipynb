# Data Scientist Work: Demand Estimation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
df=pd.read_csv('/content/analytics_final_project/demand_est_project/air_fryers_clean_brand_year.csv')
df.head()
```

```python
y=df['log_brand_share']

brand_dummies=pd.get_dummies(df['brand'],drop_first=True, prefix="brand", dtype=int)
year_dummies=pd.get_dummies(df['year'],drop_first=True, prefix="year", dtype=int)
feature_cols = [
    'compact_share',
    'dual_basket_share',
    'oven_style_share',
    'rotisserie_share',
    'window_share']

X=pd.concat([df[['avg_price','avg_rating']+ feature_cols],brand_dummies,year_dummies],axis=1)
```

```python
model=LinearRegression()
model.fit(X,y)

predicted_log_share=model.predict(X)
r2=r2_score(y,predicted_log_share)

coef_table = pd.DataFrame({
    "feature": X.columns,
    "coefficient": model.coef_})

print("R-squared:", r2)
coef_table
```

### Questions

1. What is the estimated price coefficient, $\hat{\beta}_{price}$?


```python
coef_table.loc[coef_table['feature']=='avg_price']
```

The estimated price coefficient is -0.037668.

2. Is it negative? Why is that important?


The price coefficient is negative, which means that as the price increases, the demand decreases. This suggests that consumers are less likely to buy more expensive air fryers. This is important for companies selecting the appropriate price for their product, since higher prices can lead to lower demand.

3. Which product features are associated with higher demand?


```python
coef_table.sort_values(by='coefficient', ascending=False)
```

The two strongest positive predictors of demand are window_share (12.88) and compact_share (9.82). This suggests that air fryers that contain windows or those that are compact are the most valued.

4. Which brand dummy coefficients are largest? Remember that these are interpreted relative to the dropped brand.





Cuisinart (6.42) has the largest dummy coefficient relative to the omitted brand. Ninja (5.84) and Instant Pot (4.63) also have a strong positive effect. All listed brands have positive coefficients, meaning that they outperform the reference brand in terms of demand, holding price, rating, and product features constant.

5. Which year dummy coefficients are largest? Remember that these are interpreted relative to the dropped year.


The year 2020 has the highest dummy coefficient (0.12), meaning that demand for air fryers was highest in 2020 relative to the dropped year. The second highest year coefficient is 2021, with a value of 0.042.

6. What is the model's $R^2$?

The model's R^2 is 0.763, meaning 76% of the variation in log brand share is explained by the independent variables. This indicates an overall strong explanatory power and suggests that price, product features, brand effects, and time effects are key drivers of demand in this market.
